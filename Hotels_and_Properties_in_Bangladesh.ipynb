```python
#Replace rating values where it is written 'Review score' with the text 'None'
df['rating_text'].replace('Review score','None',inplace=True)
df
```

```python
#Find out the data types for each column in the dataframe
df.info()
```

We can see that the 'price' and 'num_reviews' columns are considered as text values here, instead of numeric. This is because of the presence of the comma character to separate values that are 1000 or more. Moreover, the 'price' column consist of the text BDT before each number. So we need to convert these to pure numbers

```python
df[df['num_reviews']=="1,417"]
```

```python
#Convert num_reviews and price to numeric values
df['num_reviews'] = df['num_reviews'].str.replace(',', '')
df['num_reviews'] = df['num_reviews'].astype(int)
df['price'] = df['price'].str.replace(',', '')
df['price'] = df['price'].str.replace('BDT ', '')
df['price'] = df['price'].astype(int)
df.info()
```

```python
df.iloc[:25]
```

```python
#Find hotels that have number of reviews less than or equal to 5
df[df['num_reviews']<=10]
```

```python
#Drop the hotels that have number of reviews less than or equal to 5
df.drop(df[df['num_reviews'] <= 10].index, inplace = True)
df.shape
```

```python
#Replace rating values where it is -1.0 with Nan values, since they just mean that there is no rating
df['staff'].replace(-1.0,np.nan,inplace=True)
df['facilities'].replace(-1.0,np.nan,inplace=True)
df['cleanliness'].replace(-1.0,np.nan,inplace=True)
df['comfort'].replace(-1.0,np.nan,inplace=True)
df['value_for_money'].replace(-1.0,np.nan,inplace=True)
df['location_rating'].replace(-1.0,np.nan,inplace=True)
df
```

```python
#Let's take a look at counts for different types of ratings
df['rating_text'].value_counts()
```

As we can see, there's only 1 hotel that got an 'Exceptional' rating, and 102 hotels have no rating at all!

```python
#View the count,mean, sd, min, max, etc for each (numerical) column, by star rating
star_presence=df.groupby('stars')
#star_presence.first()
star_presence.describe()
```

```python
# In some places, the location is included along with "Dhaka". To keep things consistent, we want to change these to just "Dhaka"
df["city"] = df['city'].apply(lambda x: x.split(' ')[-1] if ('Dhaka' in x) else x)
df

#df['city']=df['city'].str.split(' ').str[-1]
#df
#mystr="Gulshan, Dhaka"
#mystr=mystr.split(' ')
#mystr[-1]
```

```python
#I also noticed that "Tourist Home" has "Uttara, Abdullahpur" as its location, which also basically Dhaka, so let's change that value too
df['city'].replace("Uttara, Ābdullāpur","Dhaka",inplace=True)
df
```
